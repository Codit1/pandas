```python
import pandas as pd
import numpy as np
import string
```

```python
arr = np.random.randint(0,100, (1, 20))

print(arr)

my_series_arr = pd.Series(arr[0])
print(my_series_arr)
```

```python
flatten_arr = arr.reshape(-1)

custom_index = np.random.choice(list(string.ascii_lowercase), 20)

my_series_arr = pd.Series(flatten_arr, index=custom_index)
print(my_series_arr)
```
