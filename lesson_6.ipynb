```python
import pandas as pd
import numpy as np
```

```python
my_df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')

my_df
```

```python
new_df = my_df[ : 5]
new_df
```

```python
total_order = [ 32, 46, 33, 12, 88 ]
new_df["total_order"] = total_order
new_df
```

```python
total_order_list = np.array(new_df["total_order"])
prices_list = np.array(new_df["item_price"])

prices_list = [ float(p.replace("$", "")) for p in prices_list ]

# prices_list

total_amount_of_order_items = total_order_list * prices_list

amount = total_amount_of_order_items
```

```python
new_df["total_amount_of_order_items"] = [ str(f"${a}") for a in amount ]
new_df
```
